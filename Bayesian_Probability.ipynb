**Part 2: Bayesian Probability using the IMDb Movie Reviews Dataset**

Step 1: Import the required library.
This is used to load, organize, and analyze the IMDb movie reviews dataset.



```python
import pandas as pd
```

Step 2: Load the dataset.


This loads the IMDb Movie Reviews dataset into a DataFrame named `df` so that it can be analyzed.


```python
df = pd.read_csv("IMDB Dataset.csv")
```

Step 3: Explore the dataset.
This displays the dataset's structure, including the number of rows, columns, data types, and whether there are any missing values.

```python
df.info()
```

Step 4: Count positive and negative reviews.
This counts how many reviews are labeled as positive and how many are labeled as negative.

```python
# Count the number of positive and negative reviews
df["sentiment"].value_counts()
```

Step 5: Calculate the Prior Probability.
The prior probability represents the probability that a randomly selected review is positive before considering any keyword.

```python
# Count positive reviews
positive_reviews = (df["sentiment"] == "positive").sum()

# Count the total number of reviews
total_reviews = len(df)

# Calculate the Prior Probability
prior = positive_reviews / total_reviews

print(f"Positive Reviews: {positive_reviews}")
print(f"Total Reviews: {total_reviews}")
print(f"Prior Probability P(Positive): {prior:}")
print(f"Prior Probability percentage P(Positive): {prior:.2%}")
```

Step 6: Create a dataset containing only positive reviews.
This creates a new DataFrame containing only positive reviews. It will be used when calculating the likelihood.

```python
# Select only positive reviews
positive_df = df[df["sentiment"] == "positive"]

# Display the first few positive reviews
positive_df.head()
```

Step 7: Check the size of the positive dataset.
This displays the dimensions of the positive reviews dataset to confirm how many positive reviews are available.

```python
# Display the number of rows and columns
print(positive_df.shape)
```

Step 8: Count positive reviews containing the keyword.
This counts how many positive reviews contain the keyword "excellent", ignoring uppercase and lowercase differences.

```python
keywords = ["excellent", "brilliant", "wonderful", "perfect"]
keyword = "excellent"

positive_keyword_count = positive_df["review"].str.contains(
    keyword,
    case=False,
    na=False
).sum()

print("Positive reviews containing", keyword, ":", positive_keyword_count)
```

Step 9: Calculate the Likelihood.
The likelihood is the probability of observing the keyword in a review, given that the review is already known to be positive.

```python
likelihood = positive_keyword_count / positive_reviews

print("Likelihood P(excellent | Positive):", likelihood)
```

Step 10: Count all reviews containing excellent keyword.
This counts how many reviews in the entire dataset contain the excellent keyword, regardless of their sentiment.

```python
keyword = "excellent"

keyword_count = df["review"].str.contains(
    keyword,
    case=False,
    na=False
).sum()

print("Total reviews containing", keyword, ":", keyword_count)
```

Step 11: Calculate the Marginal Probability.
The marginal probability is the probability that any randomly selected review contains the keyword.

```python
marginal = keyword_count / total_reviews

print("Marginal Probability P(excellent):", marginal)
```

Step 12: Apply Bayes' Theorem.
This applies Bayes' Theorem to calculate the probability that a review is positive given that it contains the keyword.

```python
posterior = (likelihood * prior) / marginal

print("Posterior Probability P(Positive | excellent):", posterior)
```

Step 13: Generalize the Bayesian Calculation for Multiple Keywords.

This section creates a reusable function that calculates the prior, likelihood, marginal, and posterior probabilities for any keyword. The function is then applied to multiple positive keywords, and the results are displayed in a table for easy comparison.

```python
# ============================================
#  Bayesian Probability for Multiple Keywords
# ============================================

# List of positive keywords
keywords = ["excellent", "brilliant", "wonderful", "perfect"]

# Prior Probability
positive_reviews = (df["sentiment"] == "positive").sum()
total_reviews = len(df)
prior = positive_reviews / total_reviews

# Create a DataFrame containing only positive reviews
positive_df = df[df["sentiment"] == "positive"]


def bayes_probability(keyword):
    """
    Computes Prior, Likelihood, Marginal and Posterior
    for a given keyword.
    """

    # Number of positive reviews containing the keyword
    positive_keyword_count = positive_df["review"].str.contains(
        keyword,
        case=False,
        na=False
    ).sum()

    # Likelihood P(keyword | Positive)
    likelihood = positive_keyword_count / positive_reviews

    # Number of ALL reviews containing the keyword
    keyword_count = df["review"].str.contains(
        keyword,
        case=False,
        na=False
    ).sum()

    # Marginal P(keyword)
    marginal = keyword_count / total_reviews

    # Posterior P(Positive | keyword)
    posterior = (likelihood * prior) / marginal if marginal != 0 else 0

    return {
        "Keyword": keyword,
        "Prior": round(prior, 4),
        "Likelihood": round(likelihood, 4),
        "Marginal": round(marginal, 4),
        "Posterior": round(posterior, 4)
    }


# Calculate probabilities for all keywords
results = []

for keyword in keywords:
    results.append(bayes_probability(keyword))

# Display results as a table
results_df = pd.DataFrame(results)

print("Bayesian Probability Results")
display(results_df)
```

**Interpretation of Results**

The posterior probability shows the likelihood that a movie review is positive after observing a specific keyword. Keywords with higher posterior probabilities are stronger indicators of positive sentiment. This demonstrates how Bayes' Theorem updates our initial belief (the prior probability) by incorporating evidence from the review text (the keyword). The results can help identify which words are the strongest predictors of positive movie reviews.
